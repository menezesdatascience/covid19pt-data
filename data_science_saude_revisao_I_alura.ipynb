```
# o .iloc[:, -1] poderá ser utilizado no lugar de [last_date]; com a exibição de todos os valores iguais:
# https://stackoverflow.com/questions/40144769/how-to-select-the-last-column-of-dataframe

def latest_by_country (data):
  return data.groupby('Country/Region').sum().iloc[:, -1]

latest_by_country(confirmed)
latest_by_country(recovered)
latest_by_country(deaths)
```

```
informations = [latest_by_country(confirmed),latest_by_country(recovered),latest_by_country(deaths)]
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'recovered', 'deaths']
combined
```

```
# Taxa de letalidade será expressa em %
# o número de casos letais dentro dos casos letais e não letais
# https://www.who.int/emergencies/diseases/novel-coronavirus-2019

sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
letality_rate_2 = sum_up['deaths'] / (sum_up['deaths'] + sum_up['recovered'])
print (letality_rate_1 * 100)
print (letality_rate_2 * 100)
```

```
# https://pt.stackoverflow.com/questions/176243/como-limitar-n%C3%BAmeros-decimais-em-python

round(letality_rate_1, 8)*100
```
